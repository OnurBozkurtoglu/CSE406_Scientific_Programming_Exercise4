<h3> Exercise 9.6: Nearest neighbor </h3>
<h4> Write a function that takes a value z and an array A and finds the element in A that is closest to z. The
function should return the closest value, not index.</h4>

```python
def nearist_neighbor(A, z):
    return A[np.argmin(np.abs(A-z))]

A = np.random.rand(m)
print("A: ")
print(A)
z = 1
print("Nearist neighbor: ")
print(nearist_neighbor(A, z))
```
